# LandUse prediction

```python
from blocksnet.utils.service_types import SERVICE_TYPES

print("Service_types = ", len(SERVICE_TYPES))
```

```python
landuse_to_service = dict()

for i in SERVICE_TYPES:
    for j in i['land_use']:
        if j in landuse_to_service.keys():
            landuse_to_service[j].add(i['name'])
        else:
            landuse_to_service[j] = {i['name']}


for i in landuse_to_service.items():
    print(f"land use: {i[0]};  services: {i[1]}")
```

```python
for i in landuse_to_service.items():
    print(f"land use: {i[0]};  services: {len(i[1])}")
```

```python
bricks_sum = 0
max_bricks = 0
service_to_bricksAM = dict()

for i in SERVICE_TYPES:
    bricks_sum += len(i['bricks'])
    if i['name'] in service_to_bricksAM.keys():
        print('DUB=', i['name'])
    else:
        service_to_bricksAM[i['name']] = len(i['bricks'])

print('amount bricks = ', bricks_sum)
```

```python
landuse_to_maxbricks = dict()
for i in landuse_to_service.items():
    max_bricks_ = 0
    for j in i[1]:
        max_bricks_ += service_to_bricksAM[j]
    landuse_to_maxbricks[i[0]] = max_bricks_
    print(f"land use: {i[0]}  --  max_bricks: {landuse_to_maxbricks[i[0]]}")
```
