# Development indicators

```python
import pandas as pd

blocks_gdf = pd.read_pickle('./../../data/saint_petersburg/blocks.pickle')
```

```python
from blocksnet.analysis.indicators import calculate_density_indicators 

density_df = calculate_density_indicators(blocks_gdf)
density_df.head()
```

```python
from blocksnet.analysis.indicators import calculate_development_indicators 

development_df = calculate_development_indicators(density_df)
development_df.head()
```

```python
blocks_gdf[['geometry']].join(development_df).plot('footprint_area', legend=True, figsize=(10,8)).set_axis_off()
```
