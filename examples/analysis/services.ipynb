```python
import pandas as pd

blocks = pd.read_pickle('./../data/blocks.pickle')
blocks.head(1)
```

```python
services_dfs = {}

for column in blocks.columns:
    if 'buildings' in column:
        continue
    if 'count_' not in column:
        continue
    service_type = column.removeprefix('count_')
    services_dfs[service_type] = blocks[[column]].rename(columns={column:'count'})
```

```python
from blocksnet.analysis.services import collocation_matrix

collocation_mx = collocation_matrix(services_dfs)
collocation_mx.head()
```
