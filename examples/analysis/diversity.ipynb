```python
import pandas as pd
import geopandas as gpd

blocks =  pd.read_pickle('./../data/saint_petersburg/blocks.pickle')
```

```python
from blocksnet.analysis.diversity import shannon_diversity

diversity = shannon_diversity(blocks)
diversity.head()
```

```python
blocks[['geometry']].join(diversity).plot('shannon_diversity', legend=True, figsize=(10,8)).set_axis_off()
```
