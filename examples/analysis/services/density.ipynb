# Services density

Services density indicates the amount of services per square meter.

```python
import pandas as pd

blocks_gdf = pd.read_pickle('./../../data/saint_petersburg/blocks.pickle')
```

```python
from blocksnet.analysis.services import services_density

density_df = services_density(blocks_gdf)
density_df.head()
```

```python
blocks_gdf[['geometry']].join(density_df).plot('density', figsize=(10,8), legend=True, vmax=0.001).set_axis_off()
```
